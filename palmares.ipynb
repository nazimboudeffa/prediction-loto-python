```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "data/palmares/in.csv"

df = pd.read_csv(file, delimiter=",")
```

```python
# 2. Convertir la colonne "Dernière sortie" en datetime
df["Dernière sortie"] = pd.to_datetime(df["Dernière sortie"], format="%d/%m/%y")

# 3. Afficher un aperçu des données
print("Aperçu des données :")
print(df.head())

# 4. Statistiques descriptives
print("\nStatistiques sur 'Nombre de sorties' :")
print(df["Nombre de sorties"].describe())

print("\nStatistiques sur 'Pourcentage de sorties' :")
print(df["Pourcentage de sorties"].describe())

# 5. Corrélation entre le nombre de sorties et le pourcentage
correlation = df["Nombre de sorties"].corr(df["Pourcentage de sorties"])
print(f"\nCorrélation entre 'Nombre de sorties' et 'Pourcentage de sorties' : {correlation:.2f}")

# 6. Histogramme du nombre de sorties
plt.figure(figsize=(10, 6))
plt.hist(df["Nombre de sorties"], bins=15, color="skyblue", edgecolor="black")
plt.title("Distribution du Nombre de sorties")
plt.xlabel("Nombre de sorties")
plt.ylabel("Fréquence")
plt.grid(True)
plt.show()

# 7. Dernières sorties (top 5 les plus récentes)
print("\nTop 5 des numéros avec les sorties les plus récentes :")
print(df.sort_values("Dernière sortie", ascending=False).head())

# 8. Évolution des sorties par numéro
plt.figure(figsize=(12, 6))
plt.plot(df["Numéros"], df["Nombre de sorties"], marker="o", linestyle="-", color="darkblue")
plt.title("Nombre de sorties par Numéro")
plt.xlabel("Numéro")
plt.ylabel("Nombre de sorties")
plt.grid(True)
plt.show()
```

```python
# Top 5 des pourcentages de sorties les plus élevés
print("\nTop 5 des numéros avec le plus fort pourcentage de sorties :")
top5_pourcentage = df.sort_values("Pourcentage de sorties", ascending=False).head(5)
print(top5_pourcentage)
```
